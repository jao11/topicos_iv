3. Faça a integração por trapézios. Lembre que a área de um trapézio é dada por: $A=\frac{(a + b)h}{2}$

1. Imagine que tem um/a maratonista treinando, que num trecho de $5 km$ tem velocidade quase constante de $12 km/h$. Produza uma lista distância percorrida em intervalos de 1 minuto.

```python
def km_por_hora_para_km_por_min(v_hora):
  v_min = v_hora / 60
  return v_min

def lista_de_distancias(x, v):
  distancias = 0
  ds = []
  while distancias < x:
    distancias += (km_por_hora_para_km_por_min(v))
    ds.append(round(distancias, 1))
  return ds

print(lista_de_distancias(5, 12))
```

2. O/A mesmo/a atleta no seguinte trecho acelera por $200$ metros até chegar em $15 km/h$ por $2 km$. Queremos obter uma tabela ou gráfico da distância percorrida en função do tempo e o tempo total para chegar em $7 km$, faça isto utilizando listas.

```python
def km_por_hora_para_km_por_min(v_hora):
  v_min = v_hora / 60
  return v_min

def lista_de_distancias(xf, xi, v):
  distancias = xi
  ds = []
  while distancias <= xf:
    distancias += (km_por_hora_para_km_por_min(v))
    ds.append((distancias))
  return ds

def lista_de_distan_em_acel(xf, xi, vf, vi):
  distancias = xi
  ds =[]
  while distancias <= xf:
    distancias += km_por_hora_para_km_por_min(vi) + (xf - xi)*km_por_hora_para_km_por_min(vf - vi)
    ds.append(round(distancias, 1))
  return ds

def minuto_por_minuto(distancia):
    ts = []
    for i in range(len(distancia)):
        ts.append(i)
    return ts

import matplotlib.pyplot as plt

y = [0]
y.extend(lista_de_distancias(5, 0, 12))
y.extend(lista_de_distan_em_acel(5.2, 5, 15, 12))
y.extend(lista_de_distancias(7, 5.2, 15))
#print(y)
x = minuto_por_minuto(y)
#print(x)
plt.plot(x, y)
plt.xlabel('Tempo (min)')
plt.ylabel('Distância (km)')
```

```python
def posicao_pendulo():
  from numpy import cos, sin, sqrt

  print('{:^100}'.format('Informações do péndulo'))
  print('{:=^100}'.format(''))
  thetai = float(input('Insira a posição inicial:'))
  vi = float(input('Insira a velocidade inicial:'))
  tf = float(input('Insira o tempo de oscilação que você deseja no gráfico:'))
  w = sqrt(9.8 / 1)
  t = 0
  ds =[]
  while t <= tf:
    theta = ((thetai * cos(w * t)) + ((vi/w) * sin(w * t)))
    ds.append(theta)
    t += 0.01
  print('{:=^100}'.format(''))
  return ds


def segundo_por_segundo(theta):
    ts = []
    for i in range(len(theta)):
        ts.append(i)
    return ts

import matplotlib.pyplot as plt

y = []
y.extend(posicao_pendulo())

#print(y)
x = segundo_por_segundo(y)
# print(x)
plt.plot(x, y)
plt.xlabel('Tempo (s / 10²)')
plt.ylabel('Theta (rad)')
```

```python
def pendulo_numerico():
  import numpy as np
  from scipy.integrate import odeint
  import matplotlib.pyplot as plt

  def f(u,x):
    return (u[1], - (9.8 / 1) * np.sin(u[0]))
  y0 = [0, 0.3]
  xs = np.linspace(1, 10, 1000)
  us = odeint(f, y0, xs)
  ys = us[:,0]
  plt.figure(figsize=(15,6)) 
  plt.plot(xs, ys) 
  plt.xlabel('tempo(segundo)')
  plt.ylabel('$\Theta$(rad)')
  plt.title('D**2t - 2sen(t) = 0')
  plt.show()

pendulo_numerico()
```
